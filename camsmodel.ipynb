```python
import cv2
import numpy as np
from keras.models import load_model

# Load the pre-trained model
model2 = load_model('model2')

# Define the class labels
class_labels = ['bus', 'car', 'human', 'motorcycle']


# Define the index of the second camera (replace with the appropriate index)
#camera_index = 1

# Initialize the camera
camera = cv2.VideoCapture(0)

while True:
    # Read the frame from the camera
    ret, frame = camera.read()

    # Preprocess the frame
    resized_frame = cv2.resize(frame, (224, 224))
    normalized_frame = resized_frame / 255.0
    input_data = np.expand_dims(normalized_frame, axis=0)

    # Perform prediction
    predictions = model2.predict(input_data)
    predicted_class_index = np.argmax(predictions)
    predicted_class_label = class_labels[predicted_class_index]

    # Get the bounding box coordinates
    height, width, _ = frame.shape
    bounding_box = np.array([0, 0, width, height])

    # Draw the bounding box
    cv2.rectangle(frame, (bounding_box[0], bounding_box[1]), (bounding_box[2], bounding_box[3]), (0, 255, 0), 2)

    # Display the predicted class label on the frame
    cv2.putText(frame, predicted_class_label, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    cv2.imshow('Vehicle Detection', frame)

    # Break the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the camera and close windows
camera.release()
cv2.destroyAllWindows()
```
